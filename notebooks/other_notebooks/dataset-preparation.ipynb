# Sample Dataset Preparation

This notebook is a simple conversion from Open Platypus Dataset into a CSV file (with some modifications).

Find the dataset at https://huggingface.co/datasets/garage-bAInd/Open-Platypus

```python
!wget https://huggingface.co/datasets/garage-bAInd/Open-Platypus/resolve/main/data/train-00000-of-00001-4fe2df04669d1669.parquet
!mv train-00000-of-00001-4fe2df04669d1669.parquet train.parquet
```

```python
import pandas as pd
```

```python
df = pd.read_parquet('train.parquet')
```

```python
df['instruction'] = df.apply(
    lambda row: f"{row['input']} {row['instruction']}" if row['input'].strip() else row['instruction'],
    axis=1
)
```

```python
df.head()
```

```python
df.drop(columns=['input', 'data_source'], inplace=True)
```

```python
df.head()
```

```python
df = df.rename(columns={"instruction": "user", "output": "assistant"})
```

```python
df.head()
```

```python
df.to_csv('open-platypus.csv', index=False)
```
